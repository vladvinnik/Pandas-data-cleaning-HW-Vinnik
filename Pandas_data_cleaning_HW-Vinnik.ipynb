```python
#Question 3 What is the average arrival delay for flights to RDU? float q_3

flights_to_RDU = flights_data.loc[flights_data['dest'] == 'RDU'] #makes a new df with just RDU destinations
x = flights_to_RDU["arr_delay"].mean() #finds mean of the arr_delay column
print(x)
```

```python
#Question 4 What proportion of flights to SEA come from the two NYC airports (LGA and JFK)?  float q_4
flights_to_SEA = flights_data.loc[flights_data['dest'] == 'SEA']
total_origins = flights_to_SEA['origin'].value_counts(normalize = True)
print(total_origins) 
print("we can see that the proportion of flights coming to Seattle from NYC airports is 0.53%")
#so value counts tallies the number of unique values and the normalize method allows us to convert those numbers into frequencies, or proportions
```
